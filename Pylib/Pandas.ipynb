```python
def convert_people(cell):
    if cell == 'n.a.':
        return 'sam walton'
    return cell
    
def convert_price(cell):
    if cell=="n.a.":
        return 50
    return cell
    
    
df = pd.read_excel('stock_data.xlsx','Sheet2',converters = {'people': convert_people, 'price' : convert_price})
df
```

### count no. of times a word is in a string

```python
s = 'The working here is fantastic, but i have a problem with the kind of work they do here.'

s.count('work')
```

```python
l = s.split()
l_e = list(map((lambda x: x.strip(',.')),l))
l_e.count('work')
```

### OR

```python
l = s.split()


def strp(l):
    x = []    
    for i in l:
        x.append(i.strip(',.'))
    return x

l_e = strp(l)
l_e.count('work')
```
